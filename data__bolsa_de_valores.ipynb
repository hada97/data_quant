# Instalando a biblioteca quantstats

```python
!pip install quantstats

#matplotlib install
import quantstats as qs

# extend pandas functionality with metrics, etc.
#estender a funcionalidade do pandas com métricas, etc.
qs.extend_pandas()

# buscar os retornos diários de uma ação
meta = qs.utils.download_returns('META')

# mostrar proporção de nitidez
qs.stats.sharpe(meta)

# or using extend_pandas() :)
#meta.sharpe()
```

Visualize stock performance (META)

```python
qs.plots.snapshot(meta, title='Facebook Performance', show=True)

# can also be called via:
# stock.plot_snapshot(title='Facebook Performance', show=True)
```

Mostrando retorno mensal

```python
meta.monthly_returns()
```

Mostrando heatmao da META

```python
meta.plot_monthly_heatmap()
```
